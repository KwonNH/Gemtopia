```python
## appid를 이용하여 tag를 가져오는 function
def get_tags(appid):
    response = requests.get('http://steamspy.com/app/'+appid)

    soup = bs4.BeautifulSoup(response.text)
    genres = soup.select('ul.sub-menu a[href^=/genre]')
    tags = soup.select('ul.sub-menu a[href^=/tag]')

    genres = [a.attrs.get('href') for a in soup.select('ul.sub-menu a[href^=/genre]')]
    tags = [a.attrs.get('href') for a in soup.select('ul.sub-menu a[href^=/tag]')]

    set_genres = set(genres)
    set_tags = set(tags)

    list_genres = list(set_genres)
    list_tags = list(set_tags)

    df_genres = pd.DataFrame(list_genres)
    df_tags = pd.DataFrame(list_tags)
    
    result = df_genres.append(df_tags)
    
    return result
```

```python
## unicode list to string list
## 유니코드 리스트를 스트링 리스트로 변환하는 function
def uni_to_str(uni_list):
    new_list = []
    for i,val in enumerate(uni_list):
        tag = val.encode('utf8')
        new_list.append(tag)
    return new_list
```

```python
## 태그 Dictionary에서 key 값만 뽑아 리스트로 만든 후, 
## unicode 에서 string으로 변환하는 function

def dict_to_str_list(dictionary):
    new_list = []
    result_list = []
    
    new_list = dictionary.keys()# dictionary를 list로 변환(unicode 상태)
    
    for i,val in enumerate(new_list):
        tag = val.encode('utf8')
        result_list.append(tag) # unicode -> string
        
    return result_list
```

```python
## dataframe의 tag 딕셔너리를 string 리스트로
## 변환하여 return하는 function

def make_tag_list_of_df(dataframe,column):
    result_list = []
    tag_list = []
    
    for i,val in enumerate(dataframe):
        tag_list = dict_to_str_list(dataframe.iloc[i][column])
        result_list.append(tag_list)
       
    return result_list
```

```python
## 사용하고자 하는 태그만 남기는 function
def remain_tag(game_tags_list, tag_list):
    a = 0
    for i,val in enumerate(game_tags_list):
        if val not in tag_list:
            game_tags_list.remove(val)
            
    return game_tags_list
            
    
```

```python
## tag 딕셔너리를 리스트로 변환하고, 
## 그 리스트에서 선택하지 않은 태그는 제외한 다음
## 다시 tags column에 리스트의 첫 번째 원소를 대입

def final_df(dataframe, column, tag_list):
    for i, val in enumerate(dataframe):
        list1 = dict_to_str_list(dataframe.iloc[i][column]) # dictionary 를 string list로 변환
        list2 = remain_tag(list1, tag_list) # tag string list와 선택한 tag_list를 비교하여 일치하는 것만 return
        dataframe['tag'] = list2[0]
        
    return dataframe
        
```
