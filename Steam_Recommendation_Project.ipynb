```python
## appid를 이용하여 tag를 가져오는 function
def get_tags(appid):
    response = requests.get('http://steamspy.com/app/'+appid)

    soup = bs4.BeautifulSoup(response.text)
    genres = soup.select('ul.sub-menu a[href^=/genre]')
    tags = soup.select('ul.sub-menu a[href^=/tag]')

    genres = [a.attrs.get('href') for a in soup.select('ul.sub-menu a[href^=/genre]')]
    tags = [a.attrs.get('href') for a in soup.select('ul.sub-menu a[href^=/tag]')]

    set_genres = set(genres)
    set_tags = set(tags)

    list_genres = list(set_genres)
    list_tags = list(set_tags)

    df_genres = pd.DataFrame(list_genres)
    df_tags = pd.DataFrame(list_tags)
    
    result = df_genres.append(df_tags)
    
    return result
```

```python
## unicode list to string list
## 유니코드 리스트를 스트링 리스트로 변환하는 function
def uni_to_str(uni_list):
    new_list = []
    for i,val in enumerate(uni_list):
        tag = val.encode('utf8')
        new_list.append(tag)
    return new_list
```

```python
## 태그 Dictionary에서 key 값만 뽑아 리스트로 만든 후, 
## unicode 에서 string으로 변환하는 function

def dict_to_str_list(dictionary):
    new_list = []
    result_list = []
    
    new_list = dictionary.keys()# dictionary를 list로 변환(unicode 상태)
    
    for i,val in enumerate(new_list):
        tag = val.encode('utf8')
        result_list.append(tag) # unicode -> string
        
    return result_list
```

```python
def createC1(dataSet):
    C1=[]
    for transaction in dataSet:
        for item in transaction:
            if not [item] in C1:
                C1.append([item])

    C1.sort()
    return map(frozenset, C1)

```

```python

# [1,3,2,5] 를 # [1,3] [2,5] [2,3] [3,5] 를 # [2,3,5] 이런식으로 만드는 함수 def aprioriGen(Lk, k):

def aprioriGen(Lk, k):
    retList = []
    lenLk = len(Lk)
    for i in range(lenLk):
        for j in range(i+1, lenLk):
            L1 = list(Lk[i])[:k-2]; L2 = list(Lk[j])[:k-2]
            L1.sort(); L2.sort()
            if L1 == L2:
                retList.append(Lk[i] | Lk[j])

    return retList

# 특정 지지도 이상의 값들의 쌍을 찾음 
def apriori(dataset, minSupport = 0.5):
    C1 = createC1(dataset)
    D = map(set, dataset)
    L1 , supportData = scanD(D,C1,minSupport)
    L = [L1]
    k=2
    while (len(L[k-2]) > 0):
        Ck = aprioriGen(L[k-2],k)
        Lk,supK = scanD(D,Ck,minSupport) # 후보그룹을 모두 찾는다.
        supportData.update(supK)
        L.append(Lk) #이게 핵심!특정 지지도 이상의 그룹들만 L에 담는다.즉 가지치기
        k += 1
    return L, supportData


if __name__ == "__main__":
    print "apriori 알고리즘"

    dataset = top100owned_tags

    L, suppData = apriori(dataset)
    print "L:" + str(L)
    print "........................."
    print "suppData:" + str(suppData)

```
